```python
# Fitting the linear regression model to predict the price using sqft_living then calculated the R^2. 

X = df[['sqft_living']]
Y = df['price']
lm = LinearRegression()
lm
lm.fit(X,Y)
lm.score(X, Y)

ax=sns.scatterplot(x='sqft_living',y='price', data=df)
```

```python
# creating train and test sets;
    # Used train_test_split() method to create train and test sets.Test size is given as 0.3, which means 30% of the data goes into test sets
    #and traun set data contains 70% data.

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=101)
model = LinearRegression()
```

```python
# fitting the model
model.fit(X_train,Y_train)

# making predictions
predictions = model.predict(X_test)
```

```python
# model evaluation
print(
  'mean_squared_error : ', mean_squared_error(Y_test, predictions))
print(
  'mean_absolute_error : ', mean_absolute_error(Y_test, predictions))
```
